```python
plt.figure()
x=np.arange(5, 60, 5)
y=x**2
colours=["red"]*(len(x)-1)
colours.append("green")
plt.scatter(x, y, c=colours, s=10)
```

```python
#Line Graphs
```

```python
x=np.arange(5, 60, 5)
y1=x**2
y2=x**3
plt.figure()
plt.plot(x, y1, "-o", c="green")
plt.plot(x, y2, "-o", c="red")
```

```python
plt.legend(["A", "B"])
```

```python
plt.gca().fill_between(x, y1, y2, facecolor="pink", alpha=0.9)
```

```python
z=plt.gca().xaxis
for item in z.get_ticklabels():
    item.set_rotation(45)
```

```python
plt.xlabel("Time")
plt.ylabel("Space")
plt.title("My Plot")
```

```python
#Bar Charts
```

```python
plt.figure()
x=np.arange(5, 60, 5)
y1=x
y2=x*2
plt.bar(x, y1, width=2, color="orange")
plt.bar(x, y2, width=2, color="green", bottom=y1)
```

```python
plt.figure()
x=np.arange(5, 60, 5)
y1=x
y2=x*2
plt.barh(x, y1, height=2, color="orange")
plt.barh(x, y2, height=2, color="green", left=y1)
```

```python
plt.legend(["A", "B"])
```

```python
z=plt.gca().xaxis
for item in z.get_ticklabels():
    item.set_rotation(45)
```
