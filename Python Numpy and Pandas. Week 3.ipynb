```python
import pandas as pd
import numpy as np
```

```python
# Merging of DataFrames
```

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])
print(staff_df)
print()
print(student_df)
```

```python
val=pd.merge(staff_df, student_df, how="outer", left_on="Name", right_on="Name")
val
```

```python
doc=pd.merge(staff_df, student_df, how="outer", left_index=True, right_index=True)
doc          # in case of repetation _x and _y comes _x for left df and _y for right df
```

```python
staff_df = pd.DataFrame([{'First Name': 'Kelly', 'Last Name': 'Desjardins', 'Role': 'Director of HR'},
                         {'First Name': 'Sally', 'Last Name': 'Brooks', 'Role': 'Course liasion'},
                         {'First Name': 'James', 'Last Name': 'Wilde', 'Role': 'Grader'}])
student_df = pd.DataFrame([{'First Name': 'James', 'Last Name': 'Hammond', 'School': 'Business'},
                           {'First Name': 'Mike', 'Last Name': 'Smith', 'School': 'Law'},
                           {'First Name': 'Sally', 'Last Name': 'Brooks', 'School': 'Engineering'}])
print(staff_df)
print(student_df)
```

```python
val=pd.merge(staff_df, student_df, how="outer", left_on=["First Name", "Last Name"], right_on=["First Name", "Last Name"])
val
```

```python
a=pd.read_csv("pokemon_data.txt",delimiter="\t")
a=a.set_index("Name")
a.head()
```

```python
# Apply
```

```python
def fun(doc, f):
    c=doc*f
    return(c)
a["Twice_HP"]=a["HP"].apply(fun, f=2)
a.head()
```

```python
b=a.loc[: , "HP":"Generation"].apply(max)
b
```

```python
c=a.loc[: , "HP":"Generation"].apply(fun, f=2)
c.head()
```

```python
#groupby and agg
```

```python
a.head()
```

```python
val=a.groupby("Type 1").HP.count()
val
```

```python
val1=a.groupby("Type 1").HP.agg(["max", "min", "count", "mean"])
val1
```

```python
val2=a.groupby("Type 1").agg({"HP":"max", "Speed":"min", "Defense":"count"})
val2=val2.rename(columns={"HP":"max_HP", "Speed":"min_Speed", "Defense":"count_Defense"})
val2
```
