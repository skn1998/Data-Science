```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Subplot(s)
```

```python
x=np.arange(5, 60, 5)
y1=x**1.5
y2=x**1.9
y3=x**2
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x, y1, "-o", c="red")
plt.plot(x, y2, "-o", c="blue")
plt.subplot(1, 2, 2)
plt.plot(x, y3, "-o", c="green")
```

```python
x=np.arange(5, 60, 5)
z=np.arange(5, 80, 5)
y1=x**2
y2=z**2.5
plt.figure()
ax1=plt.subplot(1, 2, 1)
plt.plot(x, y1, "-o", c="red")
ax2=plt.subplot(1, 2, 2, sharex=ax1, sharey=ax1)
plt.plot(z, y2, "-o", c="blue")
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, sharey=True)
ax2.plot(x, y1, "-o", c="red")
ax4.scatter(z, y2, c="blue", s=50)
```
