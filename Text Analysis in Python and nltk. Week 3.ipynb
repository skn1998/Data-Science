```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
```

```python
val=pd.read_excel('Amazon_Unlocked_Mobile.xlsx')
val.head()
```

```python
print(len(val.index))
```

```python
val.dropna(inplace=True)
print(len(val.index))
```

```python
val.head()
```

```python
val=val[val['Rating']!=3]
print(len(val.index))
```

```python
def fun(doc):
    if(doc>3):
        return(1)
    else:
        return(0)
val['Category']=val['Rating'].apply(fun)
val.head(10)
```

```python
def this_fun(doc):
    return(str(doc))
val['New_Reviews']=val['Reviews'].apply(this_fun)
val.head(10)
```

```python
X=val['New_Reviews']
y=val['Category']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# CountVectorizer #
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer().fit(X_train)
print(len(vect.get_feature_names()))
```

```python
X_train_vectorized=vect.transform(X_train)
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_vectorized, y_train)
pred=model.predict(vect.transform(X_test))
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```

```python
# Using min_df and ngram_range #
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(min_df=6).fit(X_train)
print(len(vect.get_feature_names()))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(min_df=6, ngram_range=(1, 3)).fit(X_train)
print(len(vect.get_feature_names()))
```

```python
# Using Tfidf (term frequency inverse document frequency) #
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer(min_df=5).fit(X_train)
```
